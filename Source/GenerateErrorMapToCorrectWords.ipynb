```python
def removeAccents(words):
    return [normalize('NFKD', word).encode('ASCII', 'ignore').decode('ASCII') for word in words]
```

```python
newmap = {}
for x in errorMap:
    word = removeAccents([x.lower()])
    correct = removeAccents([errorMap[x].lower()])[0].split(",")
    if '-' in correct[0]:
        correct = [''.join(correct[0].split('-'))]
    if word[0] != correct[0]:
        if '\'' in word[0]:
            newtuple = tuple(word[0].split('\''))
        elif '-' in word[0]:
            newtuple = tuple(''.join(word[0].split('-')))
        else:
            newtuple = tuple(word[0].split(' ')) 
        
        if len(newtuple) == 1:
            newmap[newtuple[0]] = correct[0]
```
